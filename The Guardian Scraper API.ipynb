```python
def scraper_api(date_from, number_of_pages, folder, csv_file):
    list_of_dfs = [] # container
    today = date.today() # select date from today

    for page in range(1,number_of_pages):
        url_page =f'https://content.guardianapis.com/search?from-date={date_from}&to-date={today}&order-by=newest&show-fields=bodyText&page={page}&page-size=200&api-key={MY_API_KEY}'
        time.sleep(3)
        response_page = requests.get(url_page)
        if response_page.status_code == 200:
            list_of_dfs.append(json_normalize(response_page.json()['response']['results']))


        else: 
            print('unauthorized')

    final_data_day = pd.concat(list_of_dfs) # append the dfs
    data_folder = os.path.join(os.getcwd(), folder)
    os.makedirs(data_folder, exist_ok=True)
    output_file = os.path.join(data_folder, csv_file)

    final_data_day.to_csv(output_file, index=False)
    
 
```
