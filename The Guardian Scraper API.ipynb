- This notebook contains the script of an scraper using the API of The Guardian newspaper.

```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
import time
import json
from datetime import date 
import os
from API import *
```

```python
def scraper_api(date_from, number_of_pages, folder, csv_file):
    '''
    Returns a csv file with articles from the past month

            Parameters:
                    date_from (date-time): A date time
                    number_of_pages (int): An integer
                    folder (str): An str 
                    csv_file (str): An str

           
    '''
    list_of_dfs = [] # container
    today = date.today() # select date from today

    for page in range(1,number_of_pages): # loop over the number of pages you want to scrape
        url_page =f'https://content.guardianapis.com/search?from-date={date_from}&to-date={today}&order-by=newest&show-fields=bodyText&page={page}&page-size=200&api-key={MY_API_KEY}'
        time.sleep(3) # wait time
        response_page = requests.get(url_page) # open the page
        if response_page.status_code == 200:
            list_of_dfs.append(json_normalize(response_page.json()['response']['results'])) # create a df
        else: 
            print('unauthorized')

    final_data_day = pd.concat(list_of_dfs) # append the dfs
    data_folder = os.path.join(os.getcwd(), folder) # get the path of the data folder
    os.makedirs(data_folder, exist_ok=True) # create a dict
    output_file = os.path.join(data_folder, csv_file) # create the path of the csv file

    final_data_day.to_csv(output_file, index=False) # df to csv file
    
 
```

```python
scraper_api('2020-01-22', 5, 'data', 'articles.csv') # cada pagina nos da 200 articulos so si queremos mas articulos solo hay que cambiar el 5
```
